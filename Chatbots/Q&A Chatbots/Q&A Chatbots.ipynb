## <font color='blue'>Now we are going to see how we are going to vectorise our data</font>

```python
from keras_preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
tok=Tokenizer(filters=[])
tok.fit_on_texts(vocab)
```

```python
tok.word_index
```

```python
type(tok.word_index)
```

**So every word here has been given its own unique indexes**

```python
train_story = []
train_question = []
train_answers = []

for story,question,answer in traindata:
    train_story.append(story)
    train_question.append(question)
    train_answers.append(answer)
```

```python
train_story_seq = tok.texts_to_sequences(train_story)
```

```python
train_story
```

```python
train_story_seq 
```

**You can see we got the indexes in the sequence the data is present to us and if you want to check the if indexes are<br>
matching to the correct words then you can check from above where we used `tok.word_index` for printing**

```python
def vectorize_stories(data,word_index=tok.word_index,max_story_len=max_story_len,max_ques_len=max_ques_len):
    S = []  #S is for stories
    Q = []  #Q is for questions
    A = []   #A is for answers
    
    for story, question, answer in data:
        x = [word_index[word.lower()] for word in story]
        xq = [word_index[word.lower()] for word in question]
        y = np.zeros(len(word_index) + 1)    # Index 0 is reserved so we're going to use + 1
        y[word_index[answer]] = 1
        S.append(x)
        Q.append(xq)
        A.append(y)
        
    return (pad_sequences(S, maxlen=max_story_len),pad_sequences(Q, maxlen=max_ques_len), np.array(A))
```

```python
inputs_train, queries_train, answers_train = vectorize_stories(traindata)
```

```python
inputs_test, queries_test, answers_test = vectorize_stories(testdata)
```

### <div class='alert alert-info'> Creating model </font>

```python
from keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding
from keras.layers import Input, Activation, Dense, Permute, Dropout
from keras.layers import add, dot, concatenate
from keras.layers import LSTM
```

```python
vocab_len
```

**Placeholders for inputs**

We have 2 inputs stories and questions. We use placeholders for this. **Input()** is used to instantiate a Keras tensor.

```python
input_sequence = Input((max_story_len,))
question = Input((max_ques_len,))
```

## <font color='red'>Lets build the network now</font>

**Input Encoder m**<br>
Input gets embedded to a sequence of vectors

```python
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_len,output_dim=64))
input_encoder_m.add(Dropout(0.3))
```

**Input Encoder c**<br>
embed the input into a sequence of vectors of size query_maxlen

```python
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_len,output_dim=max_ques_len))
input_encoder_c.add(Dropout(0.3))
```

**Question Encoder**<br>
embed the question into a sequence of vectors

```python
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim=vocab_len,
                               output_dim=64,
                               input_length=max_ques_len))
question_encoder.add(Dropout(0.3))
```

**Lets encode the question and sequences**<br>


```python
input_encoded_m = input_encoder_m(input_sequence)
input_encoded_c = input_encoder_c(input_sequence)
question_encoded = question_encoder(question)
```

**We use the dot product to compute the match between first input vec seq and the query**

```python
match = dot([input_encoded_m, question_encoded], axes=(2, 2))
match = Activation('softmax')(match)
```

```python
response = add([match, input_encoded_c])  
response = Permute((2, 1))(response) 
```

**Now we  concatenate the match matrix with the question vector sequence**

```python
answer = concatenate([response, question_encoded])
```

```python
answer
```

**We are going to reduce the RNN(LSTM)**

```python
answer = LSTM(32)(answer) 
```

```python
answer = Dropout(0.5)(answer)
answer = Dense(vocab_len)(answer) 
```

```python
answer = Activation('softmax')(answer)
model = Model([input_sequence, question], answer)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit([inputs_train, queries_train], answers_train,batch_size=32,epochs=150,validation_data=([inputs_test, queries_test], answers_test))
```

### Lets save the model now

```python
filename = 'chatbot_epoch_150'
model.save(filename)
```

## Evaluating the model

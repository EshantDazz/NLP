## <font color='green'>Introduction</font>

**Finite Automata(FA) is the simplest machine to recognize patterns. The finite automata or finite state machine is an abstract machine that has five elements or tuples. It has a set of states and rules for moving from one state to another but it depends upon the applied input symbol. Basically, it is an abstract model of a digital computer**

## A Finite Automata consists of the following : 

Q : Finite set of states.<br>
Σ : set of Input Symbols.<br>
q : Initial state.<br>
F : set of Final States.<br>
δ : Transition Function.

### **<div class='alert alert-info'><font color='black'>1.  Constrict a DFA that starts with 1 and ends with 0</font> </div>**

```python
from automata.fa.dfa import DFA
```

```python
dfa=DFA(
    states={'q0','q1','q2','d'},
    input_symbols={'0','1'},
    transitions={'q0':{'0':'d','1':'q1'},
                  'q1':{'0':'q2','1':'q1'},
                  'q2':{'0':'q2','1':'q1'},
                   'd':{'0':'d','1':'d'}
                },
    initial_state='q0',
    final_states={'q2'}
                   
)
```

```python
if(dfa.accepts_input('10010')):
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('111')):
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('10')):
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('01101')):
    print("Accepted")
else:
    print("Rejected")
```

### **<div class='alert alert-info'><font color='black'>2.  Build a DFA to accept Binary strings that starts or ends with “01”</font> </div>**

```python
dfa=DFA(
    states={'A','B','C','D','E','F'},
    input_symbols={'0','1'},
    transitions={'A':{'0':'B','1':'D'},
                  'B':{'0':'E','1':'C'},
                  'C':{'0':'C','1':'C'},
                   'D':{'0':'E','1':'D'},
                    'E':{'0':'E','1':'F'},
                    'F':{'0':'E','1':'D'},
                },
    initial_state='A',
    final_states={'C','F'}
                   
)
```

```python
if(dfa.accepts_input('01101')):   #ending and starting with 01
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('0110')):    #starting with 01
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('0011011')):  #Starting with 00 and ending with 11
    print("Accepted")
else:
    print("Rejected")
```

### **<div class='alert alert-info'><font color='black'>3. DFA of a string with at least two 0’s and at least two 1’s </font> </div>**

```python
dfa=DFA(
    states={'q0','q1','q2','q3','q4','q5','q6','q7','q8'},
    input_symbols={'0','1'},
    transitions={ 'q0':{'0':'q3','1':'q1'},
                  'q1':{'0':'q4','1':'q2'},
                  'q2':{'0':'q5','1':'q2'},
                  'q3':{'0':'q6','1':'q4'},
                  'q4':{'0':'q7','1':'q5'},
                  'q5':{'0':'q8','1':'q5'},
                  'q6':{'0':'q6','1':'q7'},
                  'q7':{'0':'q7','1':'q8'},
                  'q8':{'0':'q8','1':'q8'}
                },
    initial_state='q0',
    final_states={'q8'}
                   
)
```

```python
if(dfa.accepts_input('0011011')): 
    print("Accepted")
else:
    print("Rejected")
```

```python
if(dfa.accepts_input('001')): 
    print("Accepted")
else:
    print("Rejected")
```
